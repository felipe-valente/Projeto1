Pergunta: Afim  de fazer uma correlação entre a taxa de desemprego, o IDH e a emissão de CO2, foi feita essa análise. O objetivo da pesquisa se baseia em encontrar uma correlação baixa entre um alto IDH e uma alta taxa de desemprego, uma correlação alta entre a emissão de CO2 e a taxa de desemprego e nenhuma correlação entre a emissão de  CO2 e IDH.

Introdução 
Os aspectos escolhidos foram: emissão de CO2, indice de desemprego e IDH. A emissão de CO2 de um país consiste em quanto o mesmo emitiu de CO2 medido em toneladas, essa emissão se da por conta da quantidade de carros em circulação, fábricas e a própria natureza o emite,porém a  natureza não entra na conta de emissão de CO2, na verdade paises com uma grande área verde ganham dinheiro por manterem suas areas verdes. O indice de desemprego de um país é uma estatística que mostra a porcentagem da população que tem condições de trabalahar, estão procurando emprego, porém não estão no mercado de  trabalho. O IDH é uma medida comparativa usada para medir o indice de desenvolvimento de um país, essa medida leva em consideração: a expectativa de vida de uma pessoa ao nascer, o nivel de escolaridade da população(Ensino Médio, Superior e afins) e o PIB que é o produto interno bruto, que a soma  de vlor monetário e  qualquer tipo de bem produzido.
Esperando encontrar uma baixa relação entre um alto IDH e uma alta taxa de desemprego foi feita esta pesquisa, o motivo de seguir essa linha de raciocinio é que países com um PIB maior e uma alta taxa de escolaridade podem passar uma imagem de terem menos desemprego, e o objetivo era colocar essa linha de raciocínio a prova, junto com a ideia de que países com uma alta emissão de CO2 tem menos desemprego, por conta de supostamente terem mais fabricas, ou seja, mais operários, mais oportunidade de emprego. Em relação a emissão de CO2 e o IDH, a  ideia era provar que um país que produz muito CO2 não tem nenhuma relação com um país que possui um alto IDH.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados_hdi = pd.read_excel('hdi_human_development_index.xlsx')
dados_co2 = pd.read_excel('co2_emissions_tonnes_per_person.xlsx')
dados_dsm = pd.read_excel('aged_25_54_unemployment_rate_percent.xlsx')
dados_hdi.head(20)
```

```python
hdi = dados_hdi.set_index('idh')
co2 = dados_co2.set_index('co2')
dsm = dados_dsm.set_index('dsm')


#dados.sort_values(by='', ascending=False).tail(10)
```

```python
co2.head()
```

```python
hdi_2005 = hdi.loc[:,[2005]]
co2_2005 = co2.loc[:,[2005]]
dsm_2005 = dsm.loc[:,[2005]]
```

```python
jh = hdi_2005.join(co2_2005, how='inner', lsuffix='_HDI', rsuffix='_CO2')
jh.rename( columns={"2005_HDI": "HDI de 2005", "2005_CO2": "Produção de CO2 em 2005"},inplace=True)
jh = jh.join(dsm_2005, how='inner')
jh.rename(columns={2005:"Taxa de desemprego de 2005"},inplace=True)
```

```python
hdi_2005.describe()
jh.sort_values(by='HDI de 2005', ascending=False)
```

Como pode se ver na tabela 1 alguns dos países com o maoir IDH tambem tem uma baixa produção de CO2 e aparecem entre as menores emissões, dentre estes poucos tem se os que tambem possuem uma baixa taxa de desemprego, estes países são considerados os  melhores para se morar. Alguns destes são: Suíça, Noruega, Nova  Zelândia, Coreia do Sul, entre outros.

```python
co2_2005.describe()
jh.sort_values(by='Produção de CO2 em 2005',ascending=False)
```

```python
dsm_2005.describe()
jh.sort_values(by="Taxa de desemprego de 2005",ascending=True)
```

```python
print(jh['Produção de CO2 em 2005'].describe())
jh.boxplot("Produção de CO2 em 2005")
```

Como mostra o boxplot acima, percebe-se que tem-se aparentemente 2 outliers,  todavia são 3 ouliers: Estados Unidos, Canadá e Australia. No caso dos Estados Unidos, essa produção que distoa dos demais países, não tem um motivo certo, estudiosos alegam que essa alta emissão se deu por  conta de gases naturais e por  conta de suas usinas termo elétricas, no Canadá 2005 foi o ano em que por conta da queima de combustíveis, 2,1% do CO2 emitido veio do Canadá, porém por terem atingido numeros muito altos o governo começou um programa de redução da emissão de gases, que vem feito a produção de  gases reduzir notavelmente desde 2005. Na Austrália esse  número foi realmente notável por ter sido um dos maiores nos ultimos 15 anos, mesmo assim, na época o governo Australiano tinha planos de ter uma redução nesses números, o que acabou não acontecendo.

```python
print(jh['Taxa de desemprego de 2005'].describe())
jh.boxplot("Taxa de desemprego de 2005")
```

O boxplot  acima mostra uma taxa de desemprego com um outlier muito destoante dos outros dados, sendo maior do que o dobro da media, este outlier é a Polanda, que na época passava por um período difícil em seu cenário político, tendo problemas como impostos altíssimos, uma indefinição do salário minimo, o que favorecia muito para haver uma desigualdade social muito grande, formas pouco flexiveis de emprego, ou seja, havia pouca diversidade e variedade de empregos, o que favoreceu muito para uma  alta taxa de desemprego. Por  incrível que pareça essa não foi a maior taxa de desemprego deste período, mas sim 2002 quando aproximadamente 20% da  população polandesa estava desempregada.

```python
print(jh['HDI de 2005'].describe())
jh.boxplot("HDI de 2005")
```

Neste boxplot podemos ver que embora  a média de IDH  seja alta tem-se valores muitos baixos se comparado com os outros países, o outlier destes dados são as Filipinas, que é um país com uma politica um tanto autoritaria, é um país muito simples onde seu IDH nunca sofreu uma variação notavel, mas sim uma taxa quase constante  ao longo dos anos, crescendo muito vagarosamente.

```python
jh.corr('pearson')
```

Os  gráficos de disperção  e a tabela de correlação acima comprovam que embora alguns países tenham características parecidas e aspectos parecidos, os dados nos mostram que nossa melhor correlação foi entre emissão de CO2 e HDI, que curiosamente foi a correlação esperada para ser a pior, mostrando que países que produzem mais  CO2 tendem a ter um IDH mais alto do que os outros, embora tenham-se os outliers como: Suíça, Noruega, Nova Zelândia, Coreia do Sul que são os países com um alto IDH uma baixa taxa de desemprego e uma baixa emissão  de CO2. Com estes dados conclui-se que a correlação esperada entre o IDH e a taxa de desemprego não foi atingida, porém estes aspectos possuem uma correlação consideravel como pode se ver no gráfico de tendencia de HDI por taxa de desemprego, a correlação entre os CO2 e taxa de desemprego que era esperado ser a maior acabou se mostrando a pior correlação obtida, provando que a emissão de CO2 não tem uma boa correlação com a taxa de desemprego.
